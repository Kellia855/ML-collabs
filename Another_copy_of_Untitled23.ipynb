Trying to implement Gradient Descent using One Neuron and 2 data points


```python
import numpy as np

x = np.array([1, 2])
y = np.array([3, 6])

 #Random initialization of m and b
b = np.array([3,3])
m = np.array([1,1])
```

```python
def predict(x):
  return np.matmul(m,x.T) + b
```

```python

def MSE(y):
    return np.mean((predict(x) - y)**2)
print(f'the Error is equal to {MSE(y)}')
```

```python
def dC_dm(x, y):
   return 2 * np.mean(x * (predict(x) - y))

def dC_db(x, y):
  return 2 * np.mean(predict(x) - y)
```

```python
def update_parameters(m, b, x, y, learning_rate):
  m = m - learning_rate * dC_dm(x, y)
  b = b - learning_rate * dC_db(x, y)
  return m, b

# Random initialization of m and b
m = np.array([1.0, 1.0])
b = np.array([3.0, 3.0])

learning_rate = 0.1
error = []

#I want to loop 10 times and record the error
for i in range(0,11):
  print(f'the Error is equal to {MSE(y):.2f}')
  error.append(np.round(MSE(y),2))
  m, b = update_parameters(m, b, x, y, learning_rate)

error = np.array(error)
print(error)
```

```python
import matplotlib.pyplot as plt
iterations = np.arange(0,11)
plt.plot(iterations, error)
plt.title('THE LOSS CURVE (WITH LEARNING RATE)')
plt.xlabel('This axis shows Loops/steps')
plt.ylabel('This axis shows error for each step')
plt.show()
```
